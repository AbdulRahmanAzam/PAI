```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("adult.csv")

print(df.info())
print(df.head())
```

```python
continous_features = ['age','fnlwgt','education.num','capital.gain',"capital.loss",'hours.per.week']

for feature in continous_features:
    plt.figure(figsize=(6,4))
    sns.histplot(df[feature], kde=True)
    plt.title(f"Distribution of {feature}")
    plt.xlabel(feature)
    plt.show()
```

```python
categorical_features = ['workclass', 'education', 'marital.status', 'occupation', 'relationship', 'race', 'sex', 'native.country', 'income']

for feature in categorical_features:
    plt.figure(figsize=(8, 4))
    sns.countplot(y=feature, data=df,palette='viridis', hue="sex", order=df[feature].value_counts().index)
    plt.title(f'Distribution of {feature}')
    plt.show()
```

```python
continous_features = ['age','fnlwgt','education.num','capital.gain',"capital.loss",'hours.per.week']
for feature in continous_features:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[feature])
    plt.title(f"Boxplot of {feature}")
    plt.show()
```

```python
correlation_matrix = df[continous_features].corr()

plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.scatterplot(x="education.num",y="hours.per.week", hue="income", data=df, palette="coolwarm")
plt.title('Hours per week vs Education level')
plt.show()
```
