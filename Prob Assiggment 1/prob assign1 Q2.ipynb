```python
import pandas as pd
import matplotlib.pyplot as plt

# loading data sata set to python
df = pd.read_csv('heart_disease.csv')
df
```

```python

# rows and columns
df.shape
```

```python

# summary statistics
df.describe()
```

```python

# counting unique vcalues in Gender and Smoking
print(f"distinct values in the Department: {df['Gender'].nunique()}")
print(f"distinct values in the Promotion Status: {df['Smoking'].nunique()}")
```

```python

# checking for the missing values
df.isna().sum()
```

```python

# handling missing values
for col in df.columns: # numerical
    if df[col].dtype == 'float64' or df[col].dtype == 'int64':
        df[col] = df[col].fillna(df[col].mean())
    else: ## categorical
        df[col] = df[col].fillna(df[col].mode()[0])

df.isna().sum() # checking for filled
```

```python
# histogram of age
plt.hist(df['Age'], bins=25, edgecolor='black')
plt.show()
```

```python
print(df['Heart Disease Status'].unique())
df['Heart Disease Status'] = df['Heart Disease Status'].replace({'        No': 'No'})

# Bar Plot of Heart Disease
heart_counts = df['Heart Disease Status'].value_counts()
plt.bar(heart_counts.index, heart_counts.values, color=['red','skyblue'])
plt.xlabel('Heart Disease Status')
plt.ylabel('Count')
plt.show()

```

```python
# Box plot of Cholestrol Level
plt.boxplot(df['Cholesterol Level'])
plt.show()
```
