```python
import matplotlib.pyplot as plt
import random

# Q 1
students = [i for i in range(1,21)]
heights = [random.randint(150,180) for i in range(20)]

plt.figure(figsize=(10,6))

# here tab20 is the color map which have more unique and different colors
colors = plt.cm.tab20(range(len(students)))
plt.bar(students, heights, color=colors )
plt.xlabel('Student')
plt.ylabel("Height")
plt.title("Student Heights")
plt.show()

plt.pie(heights, labels=students, colors=colors)
plt.title("Student Heights")
plt.show()
```

```python
#Question 6

math = [85, 90, 78, 88, 76, 95, 89, 92, 81, 87]
science = [80, 85, 82, 86, 79, 90, 88, 94, 78, 84]
students = [f"Student {i+1}" for i in range(10)]

plt.figure(figsize=(8, 6))
plt.scatter(math, science, color='green', label="Math vs Science")
plt.xlabel("Math Marks")
plt.ylabel("Science Marks")
plt.title(" Math vs Science Marks")
plt.legend()
plt.grid()
plt.show()
```

```python
# Question 7
years = [1900, 1920, 1940, 1960, 1980, 2000, 2020]
sea = [0.0, 0.2, 0.5, 1.0, 1.5, 2.0, 3.0]

plt.figure(figsize=(8, 6))
plt.scatter(years, sea, label="Sea Level Rise")
plt.plot(years, sea, linestyle='--', color='gray')
plt.xlabel("Years")
plt.ylabel("Sea Level Rise")
plt.show()
```

```python
#  Question 8
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 20, 25, 30]
y2 = [5, 10, 15, 20, 25]

plt.figure(figsize=(8, 6))
plt.plot(x, y1, 'o-', color='pink', label="Line Y1")
plt.plot(x, y2, 'o-', color='gray', label="Line Y2")
plt.title("Two Lines on One Graph")
plt.xlabel("Amazing X-axis")
plt.ylabel("Incredible Y-axis")
plt.legend(loc='lower right')
plt.grid()
plt.show()
```
